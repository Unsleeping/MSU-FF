```python
from scipy import special
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.linspace(90, 110, 500)
sphj_99 = []
sphj_100 = []
sphj_101 = []
```

```python
for elem in z:
    sphj_99.append(special.spherical_jn(99, elem))
    sphj_100.append(special.spherical_jn(100, elem))
    sphj_101.append(special.spherical_jn(101, elem))
```

```python
plt.ylabel('$Sph_j$')  
plt.xlabel('$z$')
plt.plot(z, sphj_99, label='$sphj_{99}$')
plt.plot(z, sphj_100, label='$sphj_{100}$')
plt.plot(z, sphj_101, label='$sphj_{99}$')
grid1 = plt.grid(True)
plt.title('Spherical Bessels functions')
plt.legend()
plt.show()
```
