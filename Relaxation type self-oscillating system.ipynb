```python
b = [.08 / 1.26, .184 / 1.24, .296 / 1.24, .36 / 1.2, 
     .368 / 1.2, .376 / 1.18, .368 / 1.2, .368 / 1.2, .392 / 1.2,
     .368 / 1.16, .344 / 1.12, .328 / 1.12, .28 / 1.12, .24 / 1.08]

f = [200, 501, 992, 1515, 1991, 2101, 2184, 2294, 2688, 3968, 5510, 6048, 8065, 10000]

# print(len(b), len(f))

log_f = []
for elem in f:
    log_f.append(math.log(elem))
    
# print(log_f)

plt.ylabel('$|beta|$')  
plt.xlabel('$f, Hz$')
plt.plot(f, b)
plt.scatter(f, b)
grid1 = plt.grid(True)
plt.title('Frequence response')
plt.show()


max_b = max(b)
print(f'|beta_max| = {max_b}')

for ind in range(len(f)):
    if b[ind] == max_b:
        f_0 = f[ind]
        break;
print(f'f_0 = {f_0} Hz')
```
